```python
import matplotlib.pyplot as plt
from bokeh.plotting import output_notebook, figure, show
import statistics

START_TIME = 1.0 # in seconds
BW = 100 # in Gbps
PORT_NO_START = 1000
```

```python
def read_flow_stats(flow_stats_filename):

  flow_stats = {}
  with open(flow_stats_filename, 'r') as f:
    for line in f:
      flow_stat_log = line.split()

      time = float(flow_stat_log[0])*1e-9 - START_TIME # in seconds
      sender = flow_stat_log[1]
      cwnd = float(flow_stat_log[4]) * 1e-3 # in KB
      rtt = float(flow_stat_log[5]) * 1e-3 # in usec

      if (rtt != 0):
        rate = cwnd * 8.0 / rtt # in Gbps
      elif (cwnd != 0):
        rate = BW

      if sender in flow_stats.keys():
        flow_stats[sender]['times'].append(time)
        flow_stats[sender]['rates'].append(rate)
        flow_stats[sender]['cwnds'].append(cwnd)
        flow_stats[sender]['rtts'].append(rtt)
      else:
        flow_stats[sender] = {}
        flow_stats[sender]['times'] = [time]
        flow_stats[sender]['rates'] = [rate]
        flow_stats[sender]['cwnds'] = [cwnd]
        flow_stats[sender]['rtts'] = [rtt]

  return flow_stats
```

```python
def read_queue_log(filename):
  retval = {}
  retval['times'] = []
  retval['q_sizes'] = []
  with open(filename, 'r') as f:
    for line in f:
      queue_log = line.split()
      try:
        time = float(queue_log[0])*1e-9 - START_TIME # in seconds
        q_size = float(queue_log[1]) * 8.0 / BW * 1e-3 # in usec
      except:
        continue

      retval['times'].append(time)
      retval['q_sizes'].append(q_size)

  return retval
```

# SWIFT - Effect of Feedback Granularity to Convergence at New Flow Arrival

```python
# granularities = [1, 30, 100, 300, 1000, 3000] # in nanoseconds
granularities = [1, 100, 500, 1000, 5000, 10000] # in nanoseconds
flow_stats_data = {}
qlog_data = {}
folder_path = 'outputs/bolt-bts-benchmarks/'
for gra in granularities:
  filename = folder_path + 'gra' + str(gra) + '_0_MSGAGG_SWIFT_PERHOP'
  flow_stats_filename = filename + '.log'
  qlog_filename = filename + '.qlen'
  flow_stats_data[str(gra)+'ns'] = read_flow_stats(flow_stats_filename)
  qlog_data[str(gra)+'ns'] = read_queue_log(qlog_filename)
```

```python
plt.figure(figsize=(8,4))
xlim = [0.0025, 0.0055]

for gra, data in qlog_data.items():
  plt.plot(data['times'],data['q_sizes'], label=gra)

if xlim is not None:
  plt.xlim(xlim)
plt.ylim([0,11])

plt.grid()
plt.title('Queue Occupancy in Response to a Flow Arrival with Different Feedback Granularities')
plt.xlabel('Time (sec)')
plt.ylabel('Queue Occupancy (usec)')
plt.legend(title='Feedback Granularity')
plt.tight_layout()

plt.show()
```

```python
output_notebook()
p = figure(plot_width=900, plot_height=400)
colors = plt.rcParams['axes.prop_cycle'].by_key()['color']
idx = 0
for gra, data in qlog_data.items():
  p.line(data['times'],data['q_sizes'], 
           legend_label=gra, line_color=colors[idx % len(colors)])
  idx += 1

show(p)
```

```python
std_devs = []
for gra in granularities:
  q_sizes = qlog_data[str(gra)+'ns']['q_sizes']
  q_sizes = q_sizes[len(q_sizes)//2:3*len(q_sizes)//4]
  std_devs.append(statistics.stdev(q_sizes))
```

```python
plt.figure()
plt.bar(qlog_data.keys(),std_devs, color=colors)
plt.xlabel('Granularity (ns)')
plt.ylabel('Queue Occupancy Variation (usec)')
plt.title('Steady State Stability of Swift by the Grannularity of the Feedback Signal')
plt.gca().set_axisbelow(True)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
